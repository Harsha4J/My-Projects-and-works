```python
#TASK1
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_excel(r"C:\Users\acer\Downloads\loan_data (2).xlsx")

# Display first 4 rows
df.head()
```

```python
#TASK2
# Shape
df.shape

# Column names
df.columns

# Data types
df.dtypes

# Detailed info
df.info()

# Statistical summary
df.describe()
```

```python
#TASK3
df.isnull().sum()

df.duplicated().sum()
df = df.drop_duplicates()

df.dtypes
```

```python
#TASK4
#Numerical variable
#Histogram
plt.figure()
sns.histplot(df['Income'],kde=True)
plt.title("Employe Income")
plt.xlabel("Income")
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplots
plt.figure()
sns.boxplot(x=df['Income'])
plt.title("Employe Income")
plt.show()
```

```python
#Categorical variables
#Value counts
#Bar plots

plt.figure()
df['Loan_Status'].value_counts().plot(kind='bar')
plt.title("Employe Loan Status")
plt.xlabel("Loan status")
plt.ylabel("Count")
plt.show()
```

```python
#Numerical vs Numerical
#Scatter plot

plt.figure()
sns.scatterplot(x='Income', y='Loan_Amount', data=df)
plt.title("Income vs Loan Amount ")
plt.show()
```

```python
#Correlation heatmap

plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Categorical vs Numerical
#Boxplot
plt.figure()
sns.boxplot(x='Loan_Status', y='Loan_Amount', data=df)
plt.title("loan amount by loan status")
plt.show()
```

```python
#TASK6
#Multivariate Analysis
sns.pairplot(df)
plt.show()

```

```python
#TASK7
#Skewness
df['Loan_Amount'].skew()
```

```python
#IQR method
Q1 = df['Loan_Amount'].quantile(0.25)
Q3 = df['Loan_Amount'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Loan_Amount'] < Q1 - 1.5*IQR) | (df['Loan_Amount'] > Q3 + 1.5*IQR)]
outliers
```

```python
#TASK
#Visualization Summary
# 1 Income
sns.histplot(df['Income'])

# 2 Sales by Category
sns.boxplot(x='Loan_Status', y='Loan_Amount', data=df)

# 3 Correlation
sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.show()
```
